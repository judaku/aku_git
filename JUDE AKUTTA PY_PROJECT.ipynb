```python
 # clearing the error
    
globe = pd.read_csv('./sand/global_food_prices.csv.zip',low_memory=False)   
globe
```

```python
# Deleting the last column without values

globe1=globe.drop(globe.columns[-1],axis=1)
globe1
```

```python
# Renaming the columns for readability

globe1.columns=['Country_ID','Country_Name','Locality_ID',
               'Locality','Market_ID','Market_Name',
               'Commodity_ID','Commodity_Name','Currency_ID',
               'Currency_Name','MarketType_ID','MarketType',
               'UnitOfMeas_ID','UnitsOfMeas','Month','Year','Price']
globe1
```

```python
# Deleting these measurement and currency id  columns since no clear-cut specification

globe2=globe1.drop(['UnitOfMeas_ID','UnitsOfMeas','Currency_ID','MarketType_ID'],axis=1)
globe2
```

```python
# To view duplicates the month column must leave since most vales associated with it are repeated

globe3=globe2.drop('Month',axis=1)
globe3
```

```python
# Show where duplicates are found

globe3.duplicated()
```

```python
# View duplicates

globe3[globe3.duplicated()]
```

```python
# Delete duplicates

globe4=globe3.drop_duplicates()
globe4
```
